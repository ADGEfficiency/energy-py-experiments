```python
from energypy.datasets import load_episodes

eps = load_episodes('./dataset/train-episodes')
```

```python
import pandas as pd
data = pd.concat(eps, axis=0)
data.columns
```

```python
data['price [$/MWh]'].plot(kind='hist')
```

```python
data['h-0-trading-price'].plot(kind='hist')
```

```python
data.plot(x='h-0-trading-price', y='price [$/MWh]', kind='scatter')
```

```python
data.plot(y='h-0-trading-price', x='price [$/MWh]', kind='scatter', figsize=(20, 20))
```

```python
import numpy as np
data['log-price'] = np.log(data['price [$/MWh]'])
data.plot(y='log-price', x='price [$/MWh]', kind='scatter', figsize=(10, 10))
```
